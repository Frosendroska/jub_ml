## Training, testing, validating

### One sample (Train)

- train, tune hyperparameters and evaluate on the whole dataset


**properties**

* very simple, fast
* model have access to the most amount of data
* doesn't provide estimate of the generalization abilities
* should not be used for the evaluation

### Hold Out (Train/Test)

- split data into two parts (usually size of the smaller size ~ 0.1-0.2)
- train classifier on the bigger part (train)
- tune hyperparameters and evaluate on the test part



**properties**

* still simple and fast
* dependens on the split => high bias
* doesn't provide estimate of the generalization abilities of the hyperparameter tuning
* usually used for the large datasets and/or heavy or costly models

```python
from sklearn.model_selection import train_test_split
```

https://scikit-learn.org/stable/modules/generated/sklearn.model_selection.train_test_split.html

<img src='imgs/im12.png'>

```python
target = train.pop('Survived')
train_x, val_x, train_y, val_y = train_test_split(train, target, train_size=0.8)
```

```python
train_x1 = train_x[['Age', 'Parch', 'Pclass']]
val_x1 = val_x[['Age', 'Parch', 'Pclass']]
```

```python
from sklearn.linear_model import LinearRegression
from sklearn.metrics import accuracy_score
```

```python
model1 = LinearRegression().fit(train_x1, train_y)
```

```python
accuracy_score(train_y, model1.predict(train_x1) > 0.5)
```

```python
accuracy_score(val_y, model1.predict(val_x1) > 0.5)
```

### Leave one out

- cross validation with k=n




**properties**
- low bias
- very computational heavy
- can't calculate some metrics
- usually done on a small dataset

```python
full = full.to_numpy()
target = target.to_numpy()
```

```python
from sklearn.model_selection import LeaveOneOut
```

```python
loo = LeaveOneOut()
```

```python
metrics = []
for train_ind, test_ind in loo.split(full, target):
    X_train, X_test = full[train_ind], full[test_ind]
    y_train, y_test = target[train_ind], target[test_ind]
    model = LinearRegression().fit(X_train, y_train)
    metrics.append(accuracy_score(y_test, model.predict(X_test) > 0.5))
```

```python
np.mean(metrics)
```
